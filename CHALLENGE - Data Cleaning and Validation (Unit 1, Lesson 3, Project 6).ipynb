```python
def subject(c):
    if ("biology" in str(c["Journal title"])) or ("blood" in str(c["Journal title"])):
        return "biology"
    elif "chem" in str(c["Journal title"]):
        return "chemistry"
    elif "psych" in str(c["Journal title"]):
        return "psychology"
    elif ("immunology" in str(c["Journal title"])) or ("infection" in str(c["Journal title"])):
        return "immunology"
    elif "physiology" in str(c["Journal title"]):
        return "physiology"
    elif "aids" in str(c["Journal title"]):
        return "AIDS"
    else:
        return "undefined"
    
df["Subject"] = df.apply(subject, axis=1)
    
```

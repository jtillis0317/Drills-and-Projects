```python
for n in range(1,100):
    if ((n % 3 == 0) & (n % 5 == 0)):
        print("fizz bang")
    elif n % 5 == 0:
        print("bang")
    elif n % 3 == 0:
        print("fizz")
    else:
        print(n)
```

```python
for n in range(1,101):
    fizz = "fizz" if (n % 3 == 0) else ""
    bang = "bang" if (n % 5 == 0) else ""
    print(f"{n} {fizz} {bang}")
```
